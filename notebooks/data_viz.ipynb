# Data Preprocessing and Cleaning for BI Dashboard

## 1. Setup

```python
import pandas as pd
import numpy as np
import os

# Define file paths
STUDENT_DATA_PATH = '../data/student_survey.xlsx'
TEACHER_DATA_PATH = '../data/teacher_burnout.xlsx'
OUTPUT_DIR = '../data/processed'

# Create output directory if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

## 2. Phase 1: Data Ingestion

### 2.1. Student Survey Dataset

```python
# Load the data and codebook sheets
student_data_df = pd.read_excel(STUDENT_DATA_PATH, sheet_name='data')
student_codebook_df = pd.read_excel(STUDENT_DATA_PATH, sheet_name='codebook')

print("Student Survey Data:")
student_data_df.head()
```

```python
print("Student Survey Codebook:")
student_codebook_df.head()
```

### 2.2. Teacher Burnout Dataset

```python
# Load the data and subtitles sheets
teacher_data_df = pd.read_excel(TEACHER_DATA_PATH, sheet_name='Data base', skiprows=2)
teacher_subtitles_df = pd.read_excel(TEACHER_DATA_PATH, sheet_name='Subtitles (codes)')

print("Teacher Burnout Data:")
teacher_data_df.head()
```
